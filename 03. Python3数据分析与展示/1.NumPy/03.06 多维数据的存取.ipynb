```python
#np.fromfile(fname,dtype=float,count=-1,sep='')
    #fname:文件、字符串
    #dtype:读取的数据类型
    #count:读入元素个数，-1表示读入整个文件
    #sep:数据分割字符串，如果是空串，写入文件为二进制
```

```python
#文本文件
a = np.arange(100).reshape(5,10,2)
a.tofile('b.dat',sep=',',format='%d')
c = np.fromfile('b.dat',dtype=np.int8,sep=',')
c
```

```python
c = np.fromfile('b.dat',dtype=np.int8,sep=',').reshape(5,10,2)
c
```

```python
#二进制文件
a = np.arange(100).reshape(5,10,2)
a.tofile('b.dat',format='%d')
c = np.fromfile('b.dat',dtype=np.int).reshape(5,10,2)
c
```

```python
#注意：a.tofile()和np.fromfile需要配合使用
    #该方法需要读取时知道存入文件时数组的维度和元素类型，可以通过元数据文件来存储额外信息
```

```python
a = np.arange(100).reshape(5,10,2)
np.save('a.npy',a)
```
