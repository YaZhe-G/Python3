```python
#.sort_index(axis=0,ascending=True)：在指定轴上根据索引进行排序，默认升序,在0轴
```

```python
import pandas as pd
import numpy as np
b = pd.DataFrame(np.arange(20).reshape(4,5),index=['c','a','d','b'])
b
```

```python
b.sort_index() #对于Series对象，只有0轴，默认一定是在0轴。所以DataFrame也是默认在0轴。默认升序
```

```python
b.sort_index(ascending=False)
```

```python
c = b.sort_index(axis=1,ascending=False) #对1轴索引进行降序排序
c
```

```python
c = c.sort_index()
c
```

```python
#.sort_values(by,axis=0,ascending=True)：在指定轴上根据数据进行排序，默认升序,在0轴
    #by:axis轴上的某个索引或索引列表
```

```python
b = pd.DataFrame(np.arange(20).reshape(4,5),index=['c','a','d','b'])
b
```

```python
c = b.sort_values(2,ascending=False)
c
```

```python
c = c.sort_values('a',axis=1,ascending=False)
c
```
