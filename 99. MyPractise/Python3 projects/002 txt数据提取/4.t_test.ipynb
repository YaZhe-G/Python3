```python
#打开并读取文件为字符串
f = open('./DATA2.txt','r')
contents = f.read()


#以换行符为分隔符得到以每一行数据为元素的列表
contents_List = contents.split('\n')
#定义一个空列表存储最终的数据
f_List = []
#以制表符为分隔符来分隔每一行中的数据
for i in contents_List:
    cow = i.split('\t')
    f_List.append(cow)
#去除最后一个空列表
f_List.pop()
#print(f_List)


#拿到0类样本和3类样本的13个特征值列表
zero_List = []
three_List = []
for i in range(len(f_List)):
    if f_List[i][1] == '0':
        f_List[i].pop(0)#去掉样本ID
        f_List[i].pop()#去掉Gender
        f_List[i].pop(0)#去掉样本类型
        zero_List.append(f_List[i])
    if f_List[i][1] == '3':
        f_List[i].pop(0)#去掉样本ID
        f_List[i].pop()#去掉Gender
        f_List[i].pop(0)#去掉样本类型
        three_List.append(f_List[i])
        
        
#把同一特征的特征值放在同一个列表中
fzero_List = []
fthree_List = []
for i in range(len(zero_List[0])):
    a = []
    b = []
    for j in range(len(zero_List)):
        a.append(float(zero_List[j][i]))
    fzero_List.append(a)
    for k in range(len(three_List)):
        b.append(float(three_List[k][i]))
    fthree_List.append(b)
print(zero_List)
#print(fthree_List)
f.close()
```
