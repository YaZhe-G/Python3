```python
card_infors = [] #全局变量：新建一个空列表来储存名片信息

def print_menu():
    '''显示系统菜单'''   #函数的文档说明
    print('#'*45)
    print('欢迎来到学生管理系统 V7.5')
    print('1. 新增学生')
    print('2. 删除学生')
    print('3. 修改学生')
    print('4. 查询学生')
    print('5. 显示所有学生')
    print('6. 保存学生信息')
    print('7. 退出系统')
    print('#'*45)

def add_students():
    '''增加一个名片信息'''
    global card_infors   #当全局变量是列表或者是字典时，global可以不写
    new_name = input('请输入学生的姓名:')
    new_qq = input('请输入学生的QQ:')
    new_email = input('请输入学生的邮箱:')
    new_address = input('请输入学生的住址:')
    #新建一个空字典来储存个人信息
    card = {} 
    card['Name'] = new_name    #card['Name']：字典的键需要加引号
    card['QQ'] = new_qq
    card['Email'] = new_email
    card['Address'] = new_address
    #将一个字典，添加到列表中
    card_infors.append(card)
    print('新增成功')

def rm_students():
    '''删除一个名片信息'''
    rm_name = input('请输入您要删除的学生姓名:')
    rm_flag = 0 #旗帜变量，防止双重输出
    for temp in card_infors:
        if temp['Name'] == rm_name:
            card_infors.remove(temp)
            print('删除成功')
            rm_flag = 1
    if rm_flag == 0:
        print('没有您要删除的学生')
        
def fix_students():
    '''修改一个名片信息'''
    fix_name = input('请输入您想要修改的学生姓名:')
    fix_flag = 0
    for temp in card_infors:
        if temp['Name'] == fix_name:
            Fnum = int(input('请输入您想要修改的学生信息的序号:[1(QQ) 2(邮箱) 3(住址)]'))
            if Fnum == 1:
                fix_qq = input('请输入您想要修改成的qq号:')
                temp['QQ'] = fix_qq
            elif Fnum == 2:
                fix_email = input('请输入您想要修改成的邮箱:')
                temp['Email'] = fix_email
            elif Fnum == 3:
                fix_address = input('请输入您想要修改成的住址:')
                temp['Address'] = fix_address
            fix_flag = 1
    if fix_flag == 0:
        print('没有您想要修改的学生的信息')
        
def find_students():
    '''查找一个名片信息'''
    find_name = input('请输入您要查询的学生姓名:')
    name_flag = 0 
    for temp in card_infors:
        if temp['Name'] == find_name:
            print('%s\t%s\t%s\t%s'%(temp['Name'],temp['QQ'],temp['Email'],temp['Address']))
            name_flag = 1
    if name_flag == 0:
        print('查无此人')
        
def show_all_students():
    '''显示所有名片信息'''
    print('Name\tQQ\tEmail\tAdress')  #表头
    for temp in card_infors:   #遍历列表中所有字典的值
        print('%s\t%s\t%s\t%s'%(temp['Name'],temp['QQ'],temp['Email'],temp['Address']))

def save_2_file():
    '''把已经添加的信息保存到文件中'''
    f = open('backup.data','w')
    f.write(str(card_infors))
    f.close()
    
def load_infor():
    '''加载之前的数据到程序中'''
    global card_infors
    #防止backup.data丢失后造成的程序异常
    try:       
        f = open('backup.data')
        card_infors = eval(f.read())   #eval():将字符串str当成有效的表达式来求值并返回计算结果
        f.close()
    except Exception:
        pass
    
def main():
    '''完成整个程序的控制'''
    #恢复(加载)之前的数据到程序中
    load_infor()
    #1. 系统界面
    print_menu()  #调用函数
    
    while 1:
        #2. 提示用户输入
        num = int(input('请输入您要执行的功能序号:'))
        #3. 执行相应的功能
        if num == 1:
            add_students()
        elif num == 2:
            rm_students()
        elif num == 3:
            fix_students()
        elif num ==4:
            find_students()
        elif num == 5:
            show_all_students()
        elif num == 6:
            save_2_file()
        elif num == 7:
            break
        else:
            print('您输入有误，请重新输入。')
            
if __name__ == '__main__':
    #调用函数
    main()
```
